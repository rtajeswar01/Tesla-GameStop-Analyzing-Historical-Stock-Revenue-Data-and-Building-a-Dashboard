### Question 2 Extracting Tesla Revenue Data Using Webscraping

```python
import requests
from bs4 import BeautifulSoup
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/58.0.3029.110 Safari/537.3"
}
```

```python
tesla_url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
response = requests.get(tesla_url, headers=headers).text
```

```python
soup = BeautifulSoup(response, "html.parser")
soup.find_all('title')
```

```python
tables = soup.find_all('table')
table_index = None
for index, table in enumerate(tables):
    if "Tesla Quarterly Revenue" in str(table):
        table_index = index
        break
if table_index is not None:
    tesla_revenue = pd.DataFrame(columns=["Date", "Revenue"])
    for row in tables[table_index].tbody.find_all("tr"):
        col = row.find_all("td")
        if col:
            Date = col[0].text
            Revenue = col[1].text.replace("$", "").replace(",", "")
            tesla_revenue = tesla_revenue.append({"Date": Date, "Revenue": Revenue}, ignore_index=True)
else:
    print("Tesla Quarterly Revenue table not found.")
```

### Question 5 & 6 - Tesla & GameStop Stock and Revenue Dashboard

```python
import plotly.express as px
```

```python
def make_graph(stock_data, revenue_data, company_name):
    fig_stock = px.line(stock_data, x=stock_data.index, y='Close', title=f"{company_name} Stock Price", labels={'Close': 'Price (USD)'})
    fig_revenue = px.line(revenue_data, x='Date', y='Revenue', title=f"{company_name} Quarterly Revenue", labels={'Revenue': 'Revenue (Millions USD)'})
    return fig_stock, fig_revenue
```

```python
# Call make_graph for Tesla
fig_tesla_stock, fig_tesla_revenue = make_graph(tesla_data, tesla_revenue, "Tesla")

# Call make_graph for GameStop
fig_gamestop_stock, fig_gamestop_revenue = make_graph(gme_data, gme_revenue, "GameStop")
```

```python
fig_tesla_stock.show()
fig_tesla_revenue.show()

fig_gamestop_stock.show()
fig_gamestop_revenue.show()
```
